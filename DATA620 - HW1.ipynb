```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
# Define empty graph
G = nx.Graph()
```

```python
# Add edges and create
G.add_edges_from([(0, 1), (0, 2), (0, 3), (2, 3), (3, 1)])
```

```python
# Initializing new inferace for matplotlib to plot
plt.figure()
```

```python
# show graph
nx.draw_networkx(G)
```
