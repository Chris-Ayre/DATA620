Identify and load a network dataset that has some categorical information available for each node.

```python
# IMPORT MODS
import pandas as pd
import networkx as nx
```

```python
# DATA IMPORT
edge_data = pd.read_csv("trump-edges.csv")
trumpgraph = nx.from_pandas_edgelist(edge_data, source="Source", target="Target")
```

```python
#verify nodes and edges were imported correctly
trumpgraph.edges()
trumpgraph.nodes()
```

### ANALYZE

__The Most Influential and Important nodes are listed below in descending order__

```python
#Most Influential
most_influential = nx.degree_centrality(trumpgraph)

for w in sorted(most_influential, key=most_influential.get, reverse=True): print(w, most_influential[w])
```

```python
#Most Important
most_important = nx.eigenvector_centrality(trumpgraph)

for w in sorted(most_important, key=most_important.get, reverse=True): print(w, most_important[w])
    
```

### CATEGORICAL DATA ANALYSIS - DEGREE CENTRALITY (MOST INFLUENTIAL)
__Looking at Special Assistants and then Family members__

```python
Spec_assist_ = edge_data[edge_data['Relationship']=='Special assistant'].copy(deep=True)
Spec_assist = nx.from_pandas_edgelist(Spec_assist_, source="Source", target="Target")

most_influential_SpecA = nx.degree_centrality(Spec_assist)

for w in sorted(most_influential_SpecA, key=most_influential_SpecA.get, reverse=True): print(w, most_influential_SpecA[w])
```

```python
family_ = edge_data[edge_data['Relationship']=='Parent/child'].copy(deep=True)
family = nx.from_pandas_edgelist(family_, source="Source", target="Target")

most_influential_fam = nx.degree_centrality(family)

for w in sorted(most_influential_fam, key=most_influential_fam.get, reverse=True): print(w, most_influential_fam[w])
```
